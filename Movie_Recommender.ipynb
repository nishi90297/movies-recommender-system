```python
movies = pd.read_csv('/Users/nishtha/Desktop/Personal/movies-recommender-system/archive/tmdb_5000_movies.csv')
credits = pd.read_csv('/Users/nishtha/Desktop/Personal/movies-recommender-system/archive/tmdb_5000_credits.csv')
```

```python
movies.head()
movies = movies.merge(credits,on='title')
movies.head(1)
```

```python
# genres
# id
# keywords
# overview
# title
# cast
# crew

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()
```

```python
movies = movies.drop_duplicates()
```

```python
import ast
def convert(obj):
    obj = ast.literal_eval(obj)  # To convert string of list into string
    l = []
    for i in obj:
        l.append(i['name'])
    return l
```

```python
movies['genres']= movies['genres'].apply(convert)
movies
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies
```

```python
import ast
def convert3(obj):
    obj = ast.literal_eval(obj)  # To convert string of list into string
    l = []
    counter = 0
    for i in obj:
        if(counter!=3):
            l.append(i['name'])
        else:
            break;
        counter +=1
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies
```

```python
import ast
def findDirector(obj):
    l = []
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
            l.append(i['name'])
            break;
    return l
```

```python
movies['crew'] = movies['crew'].apply(findDirector)
movies
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['overview'] = movies['overview'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
movies
```

```python
movies['tags'] = movies['genres'] + movies['overview'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head(5)
```

```python
movies.drop(['genres','overview','keywords','cast','crew'], axis = 1 , inplace = True)
movies
```

```python
movies['tags'] = movies['tags'].apply(lambda x : " ".join(x))
movies
```

```python
movies['tags'][0]
```

```python
movies['tags'] = movies['tags'].apply(lambda x : x.lower())
np.dtype(movies['tags'])
movies['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words = 'english' )
vector = cv.fit_transform(movies['tags']).toarray()
```

```python
vector[0]
```

```python
cv.get_feature_names_out()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies['tags'] = movies['tags'].apply(stem)
```

```python
movies['tags']
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
sorted(enumerate(similarity[1360]),reverse = True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = movies[movies['title'] == movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(enumerate(distance),reverse = True, key = lambda x:x[1])[1:6]

    for i in movies_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Batman')
```
